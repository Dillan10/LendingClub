```python
def loan_vs_funded(db,tl,xl="Funded amount",yl="Returned amount"):
    loan_fulfilled=db[db['loan_status']=='Fully Paid']
    defaulted_accounts =db[db['loan_status']=='Charged Off']
    current_accounts = db[db['loan_status']=='Current']


    loan_fulfilled_percentage= len(loan_fulfilled)/len(db) *100
    defaulted_accounts_percentage= len(defaulted_accounts)/len(db) *100
    current_accounts_percentage = len(current_accounts)/ len(db) * 100
    
    fig, ax= plt.subplots(figsize=(10, 10))
    ax.set_xlabel(xl,color="green",fontsize=30)
    ax.set_ylabel(yl,color="green",fontsize=30)
    ax.set_title(tl,color="orange",fontsize=30)

    ax.scatter(x=loan_fulfilled['funded_amnt'],y=loan_fulfilled['funded_amnt_inv'],color='green')
    ax.scatter(x=current_accounts['funded_amnt'],y=current_accounts['funded_amnt_inv'],color='black')
    ax.scatter(x=defaulted_accounts['funded_amnt'],y=defaulted_accounts['funded_amnt_inv'],color="Red")

    ax.annotate(f"Green = Loan Fulfilled -> {loan_fulfilled_percentage} %",(10,35000),color="green",fontsize=15)
    ax.annotate(f"Black = Current accounts -> {current_accounts_percentage} %",(10,32250),color="black",fontsize=15)
    ax.annotate(f"Red = Defaulted accounts -> {defaulted_accounts_percentage} %",(10,30000),color="red",fontsize=15)
    ax.annotate(f"Dataset percentage = {len(db)/len(clean_data_extra) * 100} %",(10,28000),color="grey",fontsize=15)
    ax.plot()
loan_vs_funded(clean_data_extra,"Scater plot - Loan amount vs Funded amount",)
def installment_vs_funded(db):
    loan_fulfilled=db[db['loan_status']=='Fully Paid']
    defaulted_accounts =db[db['loan_status']=='Charged Off']
    current_accounts = db[db['loan_status']=='Current']


    loan_fulfilled_percentage= len(loan_fulfilled)/len(db) *100
    defaulted_accounts_percentage= len(defaulted_accounts)/len(db) *100
    current_accounts_percentage = len(current_accounts)/ len(db) * 100
    
    fig, ax= plt.subplots(figsize=(10, 10))
    ax.set_xlabel("Loan Amount",color="green",fontsize=30)
    ax.set_ylabel("Funded amount",color="green",fontsize=30)
    ax.set_title("Scater plot - Loan amount vs Funded amount")

    ax.scatter(x=loan_fulfilled['installment'],y=loan_fulfilled['funded_amnt'],color='green')
    ax.scatter(x=current_accounts['installment'],y=current_accounts['funded_amnt'],color='black')
    ax.scatter(x=defaulted_accounts['installment'],y=defaulted_accounts['funded_amnt'],color="Red")

    ax.annotate(f"Green = Loan Fulfilled -> {loan_fulfilled_percentage} %",(10,35000),color="green",fontsize=15)
    ax.annotate(f"Black = Current accounts -> {current_accounts_percentage} %",(10,32250),color="black",fontsize=15)
    ax.annotate(f"Red = Defaulted accounts -> {defaulted_accounts_percentage} %",(10,30000),color="red",fontsize=15)
    ax.annotate(f"Dataset percentage = {len(db)/len(clean_data_extra) * 100} %",(10,28000),color="grey",fontsize=15)
    ax.plot()
```

```python
clean_data_extra['int_rate_int']=clean_data_extra['int_rate'].astype('int')
temp=clean_data_extra
temp['count']=temp['int_rate'].apply(lambda x : 1)
temp=temp.groupby(by=['int_rate_int','loan_status']).sum()
temp.iloc[ (temp.index.get_level_values('int_rate_int')==5)]['count'].sum()
#temp=temp['count']
#temp[( 5, 'Charged Off')]
#temp.index
#s.index
#temp.query(' int_rate_int == 5 & loan_status != "Charged Off" ')
```

```python
def get_interest(int_rate  , user_type):
    user_count= temp.iloc[(temp.index.get_level_values('loan_status')==user_type) & (temp.index.get_level_values('int_rate_int')==int_rate)]['count'].sum()
    total_count=temp.iloc[ (temp.index.get_level_values('int_rate_int')==int_rate)]['count'].sum()
    return user_count/total_count * 100

Get_interest_charged =lambda x : get_interest(x,"Charged Off")
Get_interest_paid =lambda x : get_interest(x,"Fully Paid")
Get_interest_current =lambda x : get_interest(x,"Current")
temp.iloc[ (temp.index.get_level_values('int_rate_int'))]
x_axis=np.arange(5,26,1)
x_axis= pd.DataFrame(x_axis,columns=['interest_rates'])
x_axis['charged_off_percentage']=x_axis['interest_rates'].apply(Get_interest_charged)
x_axis['paid_percentage']=x_axis['interest_rates'].apply(Get_interest_paid)
x_axis['current_percentage']=x_axis['interest_rates'].apply(Get_interest_current)

x_axis
```

```python
fig, ax= plt.subplots(figsize=(10, 10))
ax.set_xlabel("Interest rates",color="green",fontsize=30)
ax.set_ylabel("Percentage",color="green",fontsize=30)
ax.set_title("Interest rate trend with respect to re-payment")
ax.plot(x_axis['interest_rates'],x_axis['charged_off_percentage'],color='red')
ax.plot(x_axis['interest_rates'],x_axis['paid_percentage'],color='green')
ax.plot(x_axis['interest_rates'],x_axis['current_percentage'],color='black')

ax.annotate(f"Red =Defaulted users",(10,90),color="red",fontsize=15)
ax.annotate(f"Black = Current users",(10,80),color="black",fontsize=15)
ax.annotate(f"Green = Fully paid users%",(10,70),color="green",fontsize=15)
 #   ax.annotate(f"Dataset percentage = {len(db)/len(clean_data_extra) * 100} %",(10,28000),color="grey",fontsize=15)
ax.plot()
```
